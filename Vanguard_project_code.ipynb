```python
#ANALYSE INTERFACE (METRICS)

#Question 3: What is the completion rate (proportion of users who reach the final confirmation step) of each interface?


# Filtrer pour l'étape finale 'confirm'
confirm_step = 'confirm'

# Calculer le taux d'achèvement pour le groupe 'Test'
test_clients_total = digital_footprints_df[digital_footprints_df['Variation'] == 'Test']['client_id'].nunique()
test_clients_confirm = digital_footprints_df[(digital_footprints_df['Variation'] == 'Test') &
                                             (digital_footprints_df['process_step'] == confirm_step)]['client_id'].nunique()
test_completion_rate = test_clients_confirm / test_clients_total

# Calculer le taux d'achèvement pour le groupe 'Control'
control_clients_total = digital_footprints_df[digital_footprints_df['Variation'] == 'Control']['client_id'].nunique()
control_clients_confirm = digital_footprints_df[(digital_footprints_df['Variation'] == 'Control') &
                                                (digital_footprints_df['process_step'] == confirm_step)]['client_id'].nunique()
control_completion_rate = control_clients_confirm / control_clients_total

# Créer un DataFrame pour les résultats
completion_data = pd.DataFrame({
    'Group': ['Test', 'Control'],
    'Completion Rate': [test_completion_rate, control_completion_rate]
})

# Afficher les résultats
print(f"Total unique 'Test' clients: {test_clients_total}")
print(f"'Test' clients who reached the final step: {test_clients_confirm}")
print(f"Proportion of 'Test' clients who reached the final step: {test_completion_rate:.2%}")

print(f"Total unique 'Control' clients: {control_clients_total}")
print(f"'Control' clients who reached the final step: {control_clients_confirm}")
print(f"Proportion of 'Control' clients who reached the final step: {control_completion_rate:.2%}")

# Créer un graphique à barres pour les taux d'achèvement
plt.figure(figsize=(8, 6))
sns.barplot(x='Group', y='Completion Rate', data=completion_data, palette='viridis')
plt.title('Completion Rate by Group')
plt.xlabel('Group')
plt.ylabel('Completion Rate')
plt.ylim(0, 1)

# Ajouter les valeurs au-dessus des barres
for index, value in enumerate(completion_data['Completion Rate']):
    plt.text(index, value, f'{value:.2%}', ha='center', va='bottom')

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Merging the dataframes on 'client_id'
merged_total = pd.merge(digital_footprints_df, client_profiles_df, on='client_id', how='left')

# Ensure that 'merged_total' is sorted by 'date_time' and 'visit_id'
merged_total = merged_total.sort_values(by=['date_time', 'visit_id'])

# Add a column transforming the steps into numbers
step_mapping = {
    'start': 0,
    'step_1': 1,
    'step_2': 2,
    'step_3': 3,
    'confirm': 4
}
merged_total['process_num'] = merged_total['process_step'].map(step_mapping).fillna(np.nan)

# Detect errors (steps where the client went backward)
merged_total['prev_step'] = merged_total.groupby('visit_id')['process_num'].shift(1)
merged_total['reversion'] = merged_total['process_num'] < merged_total['prev_step']
merged_total['is_error'] = merged_total.groupby('visit_id')['reversion'].shift(-1)

# Calculate error rates
erreur_per_step = merged_total.groupby(['Variation', 'process_step'])['is_error'].sum()
total_transitions = merged_total.groupby(['Variation', 'process_step'])['is_error'].size()
error_rate_per_step = erreur_per_step / total_transitions

# Create a DataFrame to display error rates
error_rates_df = pd.DataFrame({
    'Errors': erreur_per_step,
    'Transitions': total_transitions,
    'Error Rate': error_rate_per_step
}).reset_index()

# Filter the data to keep only the 'Test' and 'Control' variations
filtered_error_rates_df = error_rates_df[error_rates_df['Variation'].isin(['Test', 'Control'])]

# Ensure 'Error Rate' is numeric
filtered_error_rates_df['Error Rate'] = pd.to_numeric(filtered_error_rates_df['Error Rate'], errors='coerce')

# Drop any rows with NaN values in 'Error Rate'
filtered_error_rates_df = filtered_error_rates_df.dropna(subset=['Error Rate'])

# Create a graph to visualize error rates
plt.figure(figsize=(12, 6))
ax = sns.barplot(x='process_step', y='Error Rate', hue='Variation', data=filtered_error_rates_df, palette='viridis')

# Annotate bars with error rates
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height:.2%}', (p.get_x() + p.get_width() / 2., height),
                ha='center', va='bottom', fontsize=10)

plt.title('Error Rate per Interface and Process Step')
plt.xlabel('Process Step')
plt.ylabel('Error Rate')
plt.xticks(rotation=45)
plt.show()

# Calculate the error rates for each group
test_error_rate = filtered_error_rates_df[filtered_error_rates_df['Variation'] == 'Test']['Error Rate']
control_error_rate = filtered_error_rates_df[filtered_error_rates_df['Variation'] == 'Control']['Error Rate']

# Perform T-test to compare error rates
t_statistic, p_value = stats.ttest_ind(test_error_rate, control_error_rate, equal_var=False)

print(f"T-Statistic: {t_statistic}, P-Value: {p_value}")

# Interpretation
if p_value < 0.05:
    conclusion = "Reject the null hypothesis. There is a significant difference between the error rates of the Test and Control groups."
else:
    conclusion = "Fail to reject the null hypothesis. There is no significant difference between the error rates of the Test and Control groups."
print(f"Conclusion: {conclusion}")

```

```python
#Question 6: How does the performance of the new design compare to the old one?
      #1) TIME PASSED KPI

#ANALYZE IF THERE IS A SIGNIFICANT DIFFERENCE BETWEEN CONTROL AND TEST

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind

# Assurez-vous que vos données sont correctement chargées
# digital_footprints_df = pd.read_csv('your_data.csv') # par exemple

# Convertir la colonne 'date_time' en format datetime
digital_footprints_df['date_time'] = pd.to_datetime(digital_footprints_df['date_time'])

# Trier les données par 'client_id', 'visit_id', et 'date_time'
digital_footprints_df = digital_footprints_df.sort_values(by=['client_id', 'visit_id', 'date_time'])

# Calculer le temps écoulé entre les étapes
digital_footprints_df['time_diff'] = digital_footprints_df.groupby(['client_id', 'visit_id'])['date_time'].diff()

# Supprimer les lignes avec des valeurs nulles dans 'time_diff'
digital_footprints_df = digital_footprints_df.dropna(subset=['time_diff'])

# Utiliser .loc pour éviter le SettingWithCopyWarning
digital_footprints_df.loc[:, 'time_diff_seconds'] = digital_footprints_df['time_diff'].dt.total_seconds()

# Calculer le temps total par visite pour chaque client
total_time_per_visit = digital_footprints_df.groupby(['client_id', 'visit_id', 'Variation'])['time_diff_seconds'].sum().reset_index()

# Séparer les données en groupes Test et Control
test_group = total_time_per_visit[total_time_per_visit['Variation'] == 'Test']['time_diff_seconds']
control_group = total_time_per_visit[total_time_per_visit['Variation'] == 'Control']['time_diff_seconds']

# Vérifier la consistance des groupes
print(f"Total visits in Test group: {len(test_group)}")
print(f"Total visits in Control group: {len(control_group)}")

# Calculer les moyennes des temps totaux pour les deux groupes
mean_test_time = test_group.mean()
mean_control_time = control_group.mean()

# Afficher les résultats
print(f"Mean total time for Test group: {mean_test_time} seconds")
print(f"Mean total time for Control group: {mean_control_time} seconds")

# Effectuer un test t indépendant pour comparer les deux groupes
t_stat, p_value = ttest_ind(test_group, control_group, equal_var=False)  # Utiliser equal_var=False si les variances ne sont pas égales

# Afficher les résultats du test t
print(f"T-Statistic: {t_stat}, P-Value: {p_value}")

# Interprétation des résultats
conclusion = "Reject the null hypothesis" if p_value < 0.05 else "Fail to reject the null hypothesis"
print(f"Conclusion: {conclusion}. There {'is' if p_value < 0.05 else 'is no'} significant difference in total time spent between Test and Control groups.")

# Déterminer quel groupe a passé le moins de temps en moyenne
if mean_test_time < mean_control_time:
    print("The Test group spent less time on average than the Control group.")
else:
    print("The Control group spent less time on average than the Test group.")

# Création du DataFrame pour le graphique
mean_times = {
    'Group': ['Test', 'Control'],
    'Mean Time (seconds)': [mean_test_time, mean_control_time]
}
mean_times_df = pd.DataFrame(mean_times)

# Créer un graphique à barres pour comparer les temps moyens
plt.figure(figsize=(10, 6))
sns.barplot(x='Group', y='Mean Time (seconds)', data=mean_times_df, palette='viridis')

# Ajouter des annotations pour les moyennes
for index, row in mean_times_df.iterrows():
    plt.text(index, row['Mean Time (seconds)'], f'{row["Mean Time (seconds)"]:.2f}', color='black', ha="center")

plt.title('Comparison of Mean Total Time Spent by Group')
plt.xlabel('Group')
plt.ylabel('Mean Time (seconds)')
plt.show()

```

```python
#Question 7: Is the difference in completion rate between the two interfaces statistically significant?
#Question 8: Is the difference in completion rate between the two interfaces profitable (threshold at 5%)?


#ANALYZE IF THERE IS A SIGNIFICANT DIFFERENCE BETWEEN CONTROL AND TEST


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Assure-toi que digital_footprints_df est chargé et prêt à l'emploi

# Filtrer pour l'étape finale 'confirm'
confirm_step = 'confirm'

# Calculer le taux d'achèvement pour le groupe 'Test'
test_clients_total = digital_footprints_df[digital_footprints_df['Variation'] == 'Test']['client_id'].nunique()
test_clients_confirm = digital_footprints_df[(digital_footprints_df['Variation'] == 'Test') &
                                             (digital_footprints_df['process_step'] == confirm_step)]['client_id'].nunique()
test_completion_rate = test_clients_confirm / test_clients_total

# Calculer le taux d'achèvement pour le groupe 'Control'
control_clients_total = digital_footprints_df[digital_footprints_df['Variation'] == 'Control']['client_id'].nunique()
control_clients_confirm = digital_footprints_df[(digital_footprints_df['Variation'] == 'Control') &
                                                (digital_footprints_df['process_step'] == confirm_step)]['client_id'].nunique()
control_completion_rate = control_clients_confirm / control_clients_total

# Afficher les taux d'achèvement pour comparaison
print(f"Taux d'achèvement du groupe 'Test': {test_completion_rate:.2%}")
print(f"Taux d'achèvement du groupe 'Control': {control_completion_rate:.2%}")

# Calculer la différence de taux d'achèvement
completion_rate_difference = test_completion_rate - control_completion_rate

# Comparer la différence au seuil de 5%
threshold = 0.05  # 5%
is_rentable = completion_rate_difference >= threshold

# Afficher les résultats
print(f"Différence de taux d'achèvement: {completion_rate_difference:.2%}")
print(f"La différence de taux d'achèvement est-elle rentable (seuil à 5%) ? {'Oui' if is_rentable else 'Non'}")

# Comparer statistiquement les taux d'achèvement (Question 7)
# Variances
p1 = test_completion_rate
n1 = test_clients_total
p2 = control_completion_rate
n2 = control_clients_total

# Variance des proportions
var_p1 = p1 * (1 - p1) / n1
var_p2 = p2 * (1 - p2) / n2

# T-statistic
t_statistic = (p1 - p2) / np.sqrt(var_p1 + var_p2)

# Degrés de liberté
df = n1 + n2 - 2

# Calculer la p-value
p_value = 1 - stats.t.cdf(abs(t_statistic), df=df)  # p-value à une queue
p_value = p_value * 2  # p-value à deux queues

print(f"T-Statistic: {t_statistic}, P-Value: {p_value}")

# Interprétation
if p_value < 0.05:
    conclusion = "Rejet de l'hypothèse nulle. Il existe une différence significative entre les taux d'achèvement des groupes Test et Control."
else:
    conclusion = "Impossible de rejeter l'hypothèse nulle. Aucune différence significative entre les taux d'achèvement des groupes Test et Control."
print(f"Conclusion: {conclusion}")

# Graphique pour visualiser les taux d'achèvement
completion_rates = {
    'Group': ['Test', 'Control'],
    'Completion Rate': [test_completion_rate, control_completion_rate]
}
completion_rates_df = pd.DataFrame(completion_rates)

# Créer un graphique à barres pour visualiser les taux d'achèvement
plt.figure(figsize=(10, 6))
sns.barplot(x='Group', y='Completion Rate', data=completion_rates_df, palette='viridis')

# Ajouter des annotations pour les taux d'achèvement
for index, row in completion_rates_df.iterrows():
    plt.text(index, row['Completion Rate'], f'{row["Completion Rate"]:.2%}', color='black', ha="center")

plt.title('Comparison of Completion Rates by Group')
plt.xlabel('Group')
plt.ylabel('Completion Rate')
plt.ylim(0, 1)
plt.show()
```
